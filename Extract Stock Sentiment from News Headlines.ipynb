```python
from selenium import webdriver
driver=webdriver.Chrome()
```

```python
driver.maximize_window()
```

```python
driver.get('https://finviz.com/news.ashx')
```

```python
from selenium.webdriver.common.by import By
news=driver.find_elements(By.XPATH, '//a[@class="nn-tab-link"]')
news_list=[]
for i in news:
    news_list.append(i.text)
print(news_list)
```

```python
time=driver.find_elements(By.XPATH, '//td[@class="text-right news_date-cell color-text is-muted"]')
time_list=[]
for i in time:
    time_list.append(i.text)
print(time_list)
```

```python
import pandas as pd
data={'Time':time_list,'News':news_list}
df=pd.DataFrame(data)
display(df)
```

```python
import re
from nltk.corpus import stopwords
from textblob import TextBlob
import nltk

# nltk.download('stopwords')

def clean_text(text):
    text = re.sub(r'[^\w\s]', '', text) 
    text = text.lower()  
    text = text.split()  
    text = [word for word in text if word not in stopwords.words('english')]
    return ' '.join(text)

df['Cleaned Reviews'] = df['News'].astype(str).apply(clean_text)

def classify_sentiment(text):
    analysis = TextBlob(text)
    return 'positive' if analysis.sentiment.polarity > 0 else 'negative' if analysis.sentiment.polarity < 0 else 'neutral'

df['Sentiment'] = df['Cleaned Reviews'].apply(classify_sentiment)
```

```python
display(df)
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['Sentiment_label']=le.fit_transform(df['Sentiment'])
```
